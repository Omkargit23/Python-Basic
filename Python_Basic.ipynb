**1. What is Python, and why is it popular?**

**2. What is an interpreter in Python?**
**Ans - ** An interpreter in Python is a program that reads and executes Python code line by line. It converts the Python code into instructions that the computer can understand and run.

Instead of translating the whole program at once, the interpreter checks each line of code and runs it immediately. This makes it easier to find and fix errors.

**3. What are pre-defined keywords in Python?**
**Ans -** Pre-defined keywords in Python are special words that have a fixed meaning in the Python programming language. These words are already reserved by Python to perform specific tasks in a program.

Because they have a special purpose, keywords cannot be used as variable names, function names, or identifiers.

Some common Python keywords are:
if, else, while, for, break, continue, def, return, import, True, False, and class.

**4. Can keywords be used as variable names?**
**Ans -** No, keywords cannot be used as variable names in Python.
Keywords are reserved words with special meanings in the language.
They are used to define the structure of a program.
Examples include if, else, for, and while, so they cannot be used as variable names.

**5. What is mutability in Python?**
**Ans -** Mutability in Python means whether an object can be changed after it is created.
If the value can be modified, it is called mutable.
Examples of mutable objects are lists, dictionaries, and sets.
If it cannot be changed, it is called immutable (like strings and tuples).

**6. Why are lists mutable, but tuples are immutable?**
**Ans -** Lists are mutable because their elements can be changed, added, or removed after creation.
This makes lists flexible for storing data that may change.
Tuples are immutable, which means their values cannot be modified once created.
Tuples are used when the data should remain constant.

**7. What is the difference between “==” and “is” operators in Python?**
**Ans-** The “==” operator is used to compare the values of two variables.
It returns True if the values are equal.

The “is” operator checks whether two variables refer to the same object in memory.
It returns True only if both variables point to the same object.

**8. What are logical operators in Python?**
Ans- Logical operators in Python are used to combine or compare conditions.
They help in making decisions in a program.
The main logical operators are and, or, and not.
They return True or False based on the conditions.

**9. What is type casting in Python?**
**Ans-** Type casting in Python means converting one data type into another.
It is used when we need a value in a different type.
For example, converting string to integer or float.
Common functions are int(), float(), and str().

**10. What is the difference between implicit and explicit type casting?**
**Ans- ** Implicit type casting is done automatically by Python when converting one data type to another.
It happens without the programmer’s instruction.

Explicit type casting is done manually by the programmer using functions like int(), float(), or str().
It is used when we want to convert a value to a specific data type.
